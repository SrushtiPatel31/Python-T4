```python
x = np.array([80,85,90,95,100,105,110,115,120,125])
y = np.array([240,250,260,270,280,290,300,310,320,330])
plt.plot(x,y)
plt.title("sports watch data",loc="right")
plt.xlabel("Avg. pulse")
plt.ylabel("calories burnage" , loc = "top")
plt.show()
```

```python
x = np.array([80,85,90,95,100,105,110,115,120,125])
y = np.array([240,250,260,270,280,290,300,310,320,330])
font1 = {'family':'serif' , 'color':'blue' , 'size':20}
font2 = {'family':'serif' , 'color':'darkred' , 'size':15}
plt.plot(x,y)
plt.title("sports watch data", fontdict = font1)
plt.xlabel("Avg. pulse" , fontdict = font2)
plt.ylabel("calories burnage" , fontdict = font2)
#plt.grid()  or plt.grid(axis='both')
#plt.show()
plt.grid(axis='y',linestyle="--")
plt.show()
```

#### subplot

```python
x = np.array([1,2,6,8])
y = np.array([3,8,1,10])
plt.subplot(1,3,1)
plt.plot(x,y)
plt.title("G-1")
plt.xlabel("x")
plt.ylabel("y")
x = np.array([1,2,6,8])
y = np.array([3,8,1,10])
plt.subplot(1,3,3)
plt.suptitle("python")
plt.plot(x,y)
plt.title("G-1")
plt.xlabel("x")
plt.ylabel("y")
# if label are overlapping then plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.array([1,2,8,6])
y = np.array([99,65,66,22])
colors = np.array(['red','green','blue','black'])
plt.scatter(x,y,color=colors)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.array([1,2,8,6])
y = np.array([45,32,31,22])
colors = np.array([50,10,20,30])
sizes = np.array([100,200,150,290])
plt.scatter(x,y,c=colors,alpha=0.5,cmap="ocean",s=sizes)
plt.colorbar()
plt.show()
```
