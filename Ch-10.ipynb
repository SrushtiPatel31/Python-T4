**Pie chart**

```python
import matplotlib.pyplot as plt
import numpy as np
plt.figure(figsize=(6,6))
x=np.array([35,25,25,45])
mylabels=["Apple","Banana","Orange","Dates"]
myexplode=[0,0.2,0,0.2]
mycolors=["red","yellow","orange","brown"]
plt.pie(x,startangle=90,labels=mylabels,explode=myexplode,shadow=True,colors=mycolors,autopct="%1.3f%%")
plt.title("Pie chart")
plt.legend(title="Four fruits")
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
plt.figure(figsize=(10,10))
x = np.array([1,2,6,8])
y = np.array([3,8,1,10])
plt.subplot(2,3,1)
plt.plot(x,y)
plt.title("plot")
plt.suptitle("Graphs")

x = np.array([1,2,6,8])
y = np.array([3,8,1,10])
plt.subplot(2,3,2)
plt.scatter(x,y)
plt.title("scatter")


x=np.array(["A","B","C","D"])
y=np.array([3,8,1,10])
plt.subplot(2,3,3)
plt.bar(x,y)
plt.title("bar")


x=np.array(["E","F","G","H"])
y=np.array([3,8,1,10])
plt.subplot(2,3,4)
plt.barh(x,y)
plt.title("barh")


x=np.array([1,1,2,3,3,4,4,5,7,8,3,8,9,10,10])
plt.subplot(2,3,5)
plt.hist(x)
plt.title("histogram")



x=np.array([35,25,25,45])
mylabels=["Apple","Banana","Orange","Dates"]
myexplode=[0,0.2,0,0.2]
mycolors=["red","yellow","orange","brown"]
plt.subplot(2,3,6)
plt.title("pie chart")


plt.pie(x,startangle=90,labels=mylabels,explode=myexplode,shadow=True,colors=mycolors,autopct="%1.3f%%")
plt.title("Pie chart")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
#creat a pie chart for the popularity data of different programming languages and display as a pie chart 
#python-29,java-19,javascript-8,c#-7,PHP-6,c,c++-5,R-3
#create and explorer view for phython and display % for each prog lang.
import matplotlib.pyplot as plt
import numpy as np
plt.figure(figsize=(8,8))
x=np.array([29,19,8,7,6,5,3])
mylabels=["python","java","javascript","c#","PHP","c,c++","R"]
myexplode=[0.2,0,0,0,0,0,0]
plt.pie(x,labels=mylabels,explode=myexplode,autopct="%1.2f%%")
plt.title("Pie chart")
plt.legend(title="Programming lang")
plt.tight_layout()
plt.show()
```

```python
#create a bar chart following data country=[USA,canada,germany,UK,france]
#GDP_per_capita=[45000,42000,52000,49000,47000]
import matplotlib.pyplot as plot
import numpy as np
x=np.array(["USA","Canada","Germany","UK","France"])
y=np.array([45000,42000,52000,49000,47000])
mycolors=["red","yellow","orange","brown","pink"]
plt.bar(x,y,color=mycolors)
plt.xlabel("Country")
plt.ylabel("GDP_per_capita")
plt.show()
```

```python
#create a line plot in which x point from 3-21 and y point according to 2x+8.
import matplotlib.pyplot as plot
import numpy as np
y=[]
x=np.arange(3,22)
for i in x:
    y.append(2*i+8)
plt.plot(x,y)
plt.show()
```
